```python
## Load text
enc='utf-8'
with open("African_helmeted_turtle.txt","r",encoding=enc) as f:
    book=f.read()

import spacy
nlp = spacy.load("en_core_web_sm")
doc=nlp(book)

def spacy_tokenizer(text):
    return [t for t in text]

tokens=spacy_tokenizer(doc)
print("TOKEN:",tokens)
```

```python
## Token attributes
def spacy_token_attr(text):
    import tabulate
    from tabulate import tabulate
    header=['token','token.i','token.is_alpha','token.is_punct','token.like_num']
    data=[(token,token.i,token.is_alpha,token.is_punct,token.like_num) for token in text]
    return tabulate(data,headers=header)

print(spacy_token_attr(doc))
```

```python
## Customize okenizer

from spacy.symbols import ORTH
nlp = spacy.blank("en")

nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"},
    {ORTH: "me"},
])
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```
